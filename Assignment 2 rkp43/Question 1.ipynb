### Problem Description

Mary and Jane are living in the same house in college, and they are often oversleep. Sometimes, when their friend Blaise comes over they stay up very late, and have a harder time getting up in the morning. Other times, Mary forgets to set her alarm, and oversleeps. For Jane her cat has a bad habit of turning off her alarm for her, thus making jane sleep in. Their professor wants to find out why Jane is late so often, and deicded to model their behaviours.

### Defining the Network

![Image representing the network](files/img/BayesNet.png)


Above is the Bayesian Network, with the following probabilities. M and J are the variables that we see, and we try to observe what A B and C are.

$A \in \{true, false\}$, corresponds that Mary set her alarm 

$B \in \{true, false\}$, corresponds that Blaise visited them last night

$C \in \{true, false\}$, corresponds that Jane's cat turned off her alarm

$M \in \{true, false\}$, corresponds that Mary woke up on time

$J \in \{true, false\}$, corresponds that Jane woke up on time

A model of this world from the perspective of Mary and Jane's professor corresponds to a probability distribution on the joint set of variables $ p(A, B, C, M, J)$ 

Through decomposition of the variables based upon the network, we get from $ P(A, B, C, M, J) $ the priors and posteriors.
$$ P(A, B, C, M, J) = P(M | A, B) P(J | B, C) P(A) P(B) P(C) $$

The base probabilities for these are as follows:

$ P(A = true) = .75 $   Mary is quite forgetful, and often does not set her alarm.

$ P(B = true) = .05 $ Blaise doesn't visit often since he lives far from them, but he does visit when he can.

$ P(C = true) = .1 $ Jane has a very smart cat, and hates the noise. Her cat has slowly gotten better at turning off her alarm. 

$ P(M = true | A = true, B = true) = .3 $ Mary may still sleep through her alarm since she was up late

$ P(M = true | A = true, B = false) = .95 $ There is some unknown cause that makes Mary still sleep in

$ P(M = true | A = false, B = true) = .1 $ Mary barely wakes up if she stays up late and forgets her alarm

$ P(M = true | A = false, B = false) = .6 $ Mary can still sometimes wake up when her alarm is off.

$ P(J = true | B = true, C = true) = .05 $ Jane has a small chance of waking up even without an alarm and staying up late

$ P(J = true | B = true, C = false) = .75 $ Jane's alarm usually can get her awake and out of bed

$ P(J = true | B = false, C = true) = .5 $ Since she has no alarm, but didn't stay up late, she might wake up on time

$ P(J = true | B = false, C = false) = 1 $ Jane gets up to her alarm as long as she slept well.

### Create the network

```python
import pymc3

#define model
model = pymc3.Model()
```

To represent the random variables, we will use the Bernoulli Distribution, as all of the variables are discrete random variables. With this, we can represent $A$, $B$, and $C$ with the following code

```python
from pymc3 import Bernoulli

with model:
    AlarmSet = Bernoulli('AlarmSet', .75)
    BlaiseVisits = Bernoulli('BlaiseVisits', .05)
    CatAlarm = Bernoulli('CatAlarm', .1)
```

We have defined the above variables so simply as they have no parents, but since Mary and Jane both have parents, we must account for those in our probability distribution for them. To do this, we use the switch function in pymc3, to do an if else representation of the outcomes of the parent variables, to determine the final state. 
$$ P(M = true | A = true, B = true) = .3 $$

$$ P(M = true | A = true, B = false) = .95 $$

$$ P(M = true | A = false, B = true) = .1 $$

$$ P(M = true | A = false, B = false) = .6 $$

$$ P(J = true | B = true, C = true) = .05 $$ 

$$ P(J = true | B = true, C = false) = .75 $$

$$ P(J = true | B = false, C = true) = .5 $$

$$ P(J = true | B = false, C = false) = 1 $$

```python
from pymc3.math import switch

with model:
    Mary = Bernoulli('Mary', switch(AlarmSet, switch(BlaiseVisits, .3, .95), switch(BlaiseVisits, .1, .6)))
    Jane = Bernoulli('Jane', switch(CatAlarm, switch(BlaiseVisits, .05, .5), switch(BlaiseVisits, .75, 1)))
```

We have now defined every variable, and so defined the model. Now we must generate samples, using the pymc3 sample function. 

```python
%matplotlib inline

nsamples = 1000

with model:
    trace = pymc3.sample(nsamples, chains = 2)
    pymc3.traceplot(trace)
    pymc3.summary(trace)
```

### Posterior Inference

The above graphs don't give us numerical values for the information that we want, and so we must calculate them ourselves, by counting the samples to get the data from the samples. 

```python
# Function created in Wet Grass notebook
# Renamed variables for my own understanding
def calcCondProb(trace, event, cond):
    for k, v in cond.items():
        # If the value for the condition is the same as in the sample, add it to the trace
        trace = [sample for sample in trace if sample[k]==v]
    # Get number of samples that match the condition
    numSamples = len(trace)
    for k, v in event.items():
        # If the values of the event are met, find those samples and add it to the trace
        trace = [sample for sample in trace if sample[k]==v]
    return len(trace)/numSamples
```

Now we can get the estimates of the probabilities for certain conditions. We will check the following posterior probabilities as examples based upon the samples generated from the code.
$$ P(C = true | J = false) $$
$$ P(B = true | J = false) $$
$$ P(B = true | M = true, J = false) $$
$$ P(B = false | M = true, J = true) $$


```python
print('P(C = true | J = false)', calcCondProb(trace, {'CatAlarm': 1},{'Jane': 0}))
print('P(B = true | J = false)', calcCondProb(trace, {'BlaiseVisits': 1},{'Jane': 0}))
print('P(B = true | M = true, J = false)', calcCondProb(trace, {'BlaiseVisits': 1},{'Mary': 1,'Jane': 0}))
print('P(B = false | M = true, J = true)', calcCondProb(trace, {'BlaiseVisits': 0},{'Mary': 1,'Jane': 1}))
print('P(M = true)', calcCondProb(trace, {'Mary': 1},{}))
print('P(A = true)', calcCondProb(trace, {'AlarmSet': 1},{}))
print('P(A = true | M = true)', calcCondProb(trace, {'AlarmSet': 1},{'Mary': 1}))
```

This returns the probilities of the hidden variables given the variables we can observe.

### Variable Elimination with a simple example
#### Part B of Question 1

For the problem, the joint probability is as follows.
$$ P(A, B, C, M, J) = P(M | A, B) P(J | B, C) P(A) P(B) P(C) $$
If we want to find the probability of $ P( A = true | M = true) $ then we must use variable elimination from the above joint probability. We start off with summing each variable that we do not know or have in what we are looking for.
$$ P( A = true | M = true) = \alpha \sum_{B, C, J}P(M = true | B, A = true)P(J | B, C) P(A = true) P(B) P(C) $$
Then we split this summation into smaller sums, breaking into each variable. Alpha represents the normalizing constant.
$$ P( A = true | M = true) = \alpha P(A) \sum_{B}P(B) P(M = true | B, A = true) \sum_{C} P(C)\sum_{J}P(J | B, C)  $$
With this, using the setup, we can eliminate $C$ and $J$ from the expression. This is because when we choose a C, we then sum over every J for a certain B and C, summing up to 1, and then 1 * P(C) for each C then returns 1, as P(C = true) + P(C = false) = 1, which then leaves behind:
$$ P( A = true | M = true) = \alpha P(A) \sum_{B}P(B) P(M = true | B, A = true) $$

We can then use this to compute and compare to the answer we found above. 
$$ P( A = true | M = true) = \alpha (.75 * ( (P(B = true) P(M = true | B = true, A = true)) + (P(B = false) P(M = true | B = false, A = true)) )) $$
$$ P( A = true | M = true) = \alpha (.75 * ( (.05 * .3) + (.95 * .95) )) $$
$$ P( A = true | M = true) = \alpha * (.75 * .9175) $$
$$ P( A = true | M = true) = \alpha * (.688125) $$

This is very close to our expected values seen above, as $\alpha = P(M = true) = .839 $ which we calculated above from the samples. This then returns the final value of 
$$Variable Elimination: P( A = true | M = true) = 0.82 $$
And our estimated value was
$$Calculated from Samples: P( A = true | M = true) = 0.844 $$
